#**Data Cleaning**
**Cleaning Data by removing special character**

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]
for char in spec_chars:
    simple_reviews_df['allSkills'] = simple_reviews_df['allSkills'].str.replace(char, ' ')
```

```python
simple_reviews_df['allSkills'] = simple_reviews_df['allSkills'].str.split().str.join(" ")
```

```python
simple_reviews_profiles = job_profiles.iloc[::, [1, 3, 8]]
simple_reviews_profiles.head(15)
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]
for char in spec_chars:
    simple_reviews_profiles['Required_Skills'] = simple_reviews_profiles['Required_Skills'].str.replace(char, ' ')

simple_reviews_profiles['Required_Skills'] = simple_reviews_profiles['Required_Skills'].str.split().str.join(" ")
simple_reviews_profiles['Required_Skills'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
from sklearn.feature_extraction.text import TfidfVectorizer
```

Importing Stopwards to clean data 

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
stoplist = set(stopwords.words("english"))
```

Using CountVectorizer function for ALLSKILLS and Required Skills data

```python
corpus = simple_reviews_df['allSkills'].values
corpus = [item for item in corpus if not isinstance(item, int)]
#vectorizer = CountVectorizer(stop_words=None)
vectorizer = CountVectorizer(stop_words = stoplist)

X = vectorizer.fit_transform(corpus)

VectorizedText=pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
VectorizedText['originalText']=pd.Series(corpus)
VectorizedText.tail(5)
```

```python
corpus_requiredskills = simple_reviews_profiles['Required_Skills'].values
corpus_requiredskills = [item for item in corpus_requiredskills if not isinstance(item, int)]
#vectorizer = CountVectorizer(stop_words=None)
vectorizer1 = CountVectorizer(stop_words = stoplist)

X = vectorizer1.fit_transform(corpus_requiredskills)

VectorizedText_profiles=pd.DataFrame(X.toarray(), columns= vectorizer1.get_feature_names())
VectorizedText_profiles['originalText']=pd.Series(corpus_requiredskills)
VectorizedText_profiles.head(5)
```

**Using TfidfVectorizer to convert the array file into Vectorized Variable **

```python
vectorizer_Tfidf = TfidfVectorizer()

trsfm=vectorizer_Tfidf.fit_transform(corpus)
Z = pd.DataFrame(trsfm.toarray(),columns=vectorizer_Tfidf.get_feature_names())
Z.head()
```

```python
vectorizer_Tfidf_profiles = TfidfVectorizer()

trsfm2=vectorizer_Tfidf_profiles.fit_transform(corpus_requiredskills)
Y = pd.DataFrame(trsfm2.toarray(),columns=vectorizer_Tfidf_profiles.get_feature_names())
Y.head()
```

# **Checking for Cosine Similarity between inverse frequency vectors creted as Y( Job Required Skills ) and Z ( Candidate Skills)**

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = []
for r in range(len(Y)):
  doc1 = Y.loc[r].values.reshape(-1, 1).tolist()
  doc2 = Z.loc[ r].values.reshape(-1, 1).tolist()
  print(doc1,doc2)
  similarities = cosine_similarity(doc1,doc2)
  cos_sim.append (similarities)

cos_sim = np.array(cos_sim)
type(cos_sim)
```
